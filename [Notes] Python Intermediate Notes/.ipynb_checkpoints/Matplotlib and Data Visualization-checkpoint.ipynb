# Matplotlib: Data Visualization 
Data Visualization is a very important aspect of Data Science, it allows us to explore the data and report the insights it provides us. **Matplotlib** is the mother of all data visualization packages, and we will be using its subpackage **pyplot** conventionally imported at **plt**.

http://matplotlib.org/api/pyplot_summary.html

```python
import matplotlib.pyplot as plt
```

Matplotlib.pyplot allows us to plot some graphs or charts to visualize our datasets! For example:  
  
## plt.plot(x,y)  
A line plot of our two lists

```python
# Our Population data (in the millions)

year = [1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
population = [2300, 2525, 3026, 3691, 4449, 5320, 6127, 6916]

# We can add that data to the plot

plt.plot(year, population) #x-axis: year, y-axis: population
plt.show() # tells python to show us the plot
```

## plt.scatter(x,y)
When you have a time scale along the horizontal axis, the line plot is your friend. But in many other cases, when you're trying to assess if there's a **correlation** between two variables, for example, the scatter plot is the better choice. Below is an example of how to build a scatter plot.

```python
# Our Population data (in the millions)
year = [1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
population = [2300, 2525, 3026, 3691, 4449, 5320, 6127, 6916]

#x-axis: year, y-axis: population, s = size list/array, c = colour, alpha = opacity
plt.scatter(year, population, s = [10,20,30,40,50,60,70,80], alpha = 0.7) 
plt.show() # tells python to show us the plot
```

## plt.scale()
Matplotlib.pyplot allows us to adjust the scale of either axis by using plt.xscale(scale) or ply.yscale(scale) and inputting one of the available scales : ‘linear’ | ‘log’ | ‘logit’ | ‘symlog’

```python
# Our Population data (in the millions)
year = [1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
population = [2300, 2525, 3026, 3691, 4449, 5320, 6127, 6916]

plt.scatter(year, population) #x-axis: year, y-axis: population
plt.yscale('log')
plt.show() # tells python to show us the plot
```

# Histograms  
A histogram is another way we can visualize our data, it allows us to see the distribution of our values along a scale. To do this we use the plt.hist() method.  
## plt.hist(x,bins)  
where x is our data, and bins is the number of bars we want on our histogram. By default, Python sets the number of bins to 10 in that case. The number of bins is pretty important. Too little bins oversimplifies reality, which doesn't show you the details. Too much bins overcomplicates reality and doesn't give the bigger picture.*see help(plt.hist)*


```python
# Let's key in some example data
values = [2,5,3,6,8,3,1,5,6,8,7,4,2,5,7,8,0,1,2,3,6,7,8,3,4,9,5,6,8,4]
plt.hist(values,bins = 10)
plt.show()
```

```python
# Build histogram with 5 bins
plt.hist(values, bins = 5)

# Show and clean up plot
plt.show()
plt.clf()

# Build histogram with 20 bins
plt.hist(values, bins = 10)

# Show and clean up again
plt.show()
plt.clf()
```

# Customizations  
There are many options we can use for Data Visualization, including different plot types and many customizations. Choice depends on both the data that you have, as well as the story you want to tell.

## Axis Labels and Title

```python
year = [1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
population = [2300, 2525, 3026, 3691, 4449, 5320, 6127, 6916]
plt.plot(year, population) , 

plt.xlabel('Year')# x-axis label
plt.ylabel('Population')# y-axis label
plt.title('World Population Projections') # Title

plt.show() # tells python to show us the plot
```

# Ticks

```python
year = [1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
population = [2300, 2525, 3026, 3691, 4449, 5320, 6127, 6916]

plt.plot(year, population) , 
plt.xlabel('Year')# x-axis label
plt.ylabel('Population')# y-axis label
plt.title('World Population Projections') # Title

plt.yticks([0,2000,4000,6000,8000,10000],
           ['0M','2000M','4000M','6000M','8000M','10000M'])

plt.show() # tells python to show us the plot
```

## Adding More Data

```python
# Adding additional data points to our graph
year = [1800,1850,1900] + year
population = [1000, 1262, 1650] + population

# We can input text by specifying a coordinate and string
plt.text(1900,5000,'1900')

plt.plot(year, population)
plt.show()
```

## Grid Lines

```python
plt.plot(year, population)
plt.grid(True) # add grid() call
plt.show()
```
