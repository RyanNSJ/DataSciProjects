# Non Linear Activation Functions
* Sigmoid function
* ReLU function
* tanh function

## 1. Sigmoid Activation Function  
* For Classification Tasks
* $S(x) = \frac{1} {1 + e^{-x}}$ - with a range from __[0,1]__

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns


def sigmoid(x):
    return (1.0/(1.0 + np.exp(-x)))

x = np.linspace(-10,10,1000)

threshold_value = 0.5
y_hat = np.zeros((1000))
y_hat[sigmoid(x)>threshold_value] = 1

threshold_value2 = 0.9
y_hat2 = np.zeros((1000))
y_hat2[sigmoid(x)>threshold_value2] = 1

# Where x_i is a predicted linear combination of the features from row i
# sigmoid(x_i) is the probability (ranging from 0 to 1) that y_i == 1
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(1,1,1)
ax.plot(x,sigmoid(x), label="Sigmoid")
ax.plot(x, y_hat, label="Threshold = 0.5")
ax.plot(x, y_hat2, label="Threshold = 0.9")
ax.legend()
plt.ylabel("$P(y_i=1|x_i)$")
plt.xlabel("$x_i$")
```

Threshold determines the required probability P(y_i=1|x_i) value for y_i to be predicted as 1. Higher Threshold values would result in more negatives and less positives predicted. i.e:
* __High Threshold__ - Predict More Negatives
    * More TN, FN
    * Less TP, FP
  
  
* __Low Threshold__ - Predict More Positives
    * More TP, FP
    * Less TN, FN

# 2. ReLU Activation Function
* __ReLU__ stands for Rectified Linear Unit 
* For Regression Tasks  
* $ReLU(x) = max(0,x)$

```python
x = np.linspace(-10,10,1000)

def relu(x):
    return np.maximum(0,x)

relu_y = relu(x)
plt.plot(x,relu_y)
```

# 3. tanh Activation Function
* __Hyperbolic Tangent Function__ - For Classification Tasks  
* $tanh(x)$ - Similiar to the Sigmoid function but going from __[-1,1]__

```python
tanh_y = np.tanh(x)
plt.plot(x,tanh_y)
```
