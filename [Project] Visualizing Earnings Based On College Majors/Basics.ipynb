## Plotting With Pandas: Earnings Based on College Majors
*How to use pyplot without using pyplot*

```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")

```

## Data Structure

First row

```python
firstrow = recent_grads.iloc[0,:]
print("First Row:\n",firstrow)
```

First and last 5 rows

```python
datahead = recent_grads.head(5)
print("First 5 Rows:\n", datahead)

datatail = recent_grads.tail(5)
print("Last 5 Rows:\n", datatail)
```

Describe each column

```python
print("Describe:\n",recent_grads.describe())
```

## Dropping Rows with Missing Data

```python
raw_data_count = recent_grads.shape[0]
print("No. of Rows (Raw) :\n",raw_data_count)
```

```python
recent_grads = recent_grads.dropna()

cleaned_data_count = recent_grads.shape[0]
print("No. of Rows (Cleaned) :\n",cleaned_data_count)
```

## Plotting Scatter plots with Panda's DataFrame.plot()

No correlation

#### No. of Women vs Median Income

```python
recent_grads.plot(x="Women", y="Median", kind="scatter", figsize=(10,10))
```

## Plotting histograms with Panda's Series.hist()  
### Sample Size  
Most Majors has a sample size below 500

```python
recent_grads["Sample_size"].hist(bins=50,range=(0,4500))
recent_grads["Sample_size"].describe()
```

### Median Income
Most jobs paid around \$36,000 a year, but a few high earning jobs pulled the mean income to \$40,000

```python
recent_grads["Median"].hist(bins=20,range=(0,80000))
recent_grads["Median"].describe()
```

### No. of graduates employed
Most Majors had under 12,000 people employed

```python
recent_grads["Employed"].hist(bins=20)
recent_grads["Employed"].describe()
```

### No. of graduates working full time

```python
recent_grads["Full_time"].hist(bins=50,range=(0,300000))
recent_grads["Full_time"].describe()
```

### % of Women taking the major
Most majors had between 33% and 70% female graduates

```python
recent_grads["ShareWomen"].hist(bins=20,range=(0,1))
recent_grads["ShareWomen"].describe()
```

### Unemployment Rate of graduates

```python
recent_grads["Unemployment_rate"].hist(bins=10,range=(0,0.1))
recent_grads["Unemployment_rate"].describe()
```

### Number of Men

```python
recent_grads["Men"].hist(bins=10,range=(0,300000))
recent_grads["Men"].describe()
```

### Number of Women

```python
recent_grads["Women"].hist(bins=10,range=(0,300000))
recent_grads["Women"].describe()
```

## Scatter Plot Matrices using Panda's scatter_matrix()

```python
from pandas.tools.plotting import scatter_matrix
```

### Sample size vs Median income

```python
scatter_matrix(recent_grads[["Sample_size","Median"]], figsize = (7,7))
```

### Sample size vs Median income vs Unemployment rate

```python
scatter_matrix(recent_grads[["Sample_size","Median","Unemployment_rate"]])
```

## Plotting Bar Charts with Panda's DataFrame.plot.bar()

```python
recent_grads_by_pay = recent_grads.sort_values("Median", ascending=False)
print(recent_grads_by_pay.iloc[:10].loc[:,["Major","Median","ShareWomen"]])

recent_grads_by_pay[:10].plot.bar(x="Median", y="ShareWomen",)
recent_grads_by_pay.iloc[-10:].plot.bar(x="Median", y="ShareWomen")

```
